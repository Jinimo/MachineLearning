# <과제1> 당뇨병 데이터 - 다중 회귀 

 - LR, SVR, RFR, XGR

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes # 당뇨병 데이터
diabetes = load_diabetes()
df  = pd.DataFrame(diabetes.data, columns = diabetes.feature_names )
df['target'] = diabetes.target
df.head()
```

```python
# train /test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
    diabetes.data, diabetes.target, test_size=0.1, random_state=2022
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score , mean_squared_error
# 경고창 무시 
import warnings 
warnings.filterwarnings('ignore')
```

```python
A = [LinearRegression, SVR, RandomForestRegressor, XGBRegressor ]
P = []

for i, a in enumerate(A):
  if i == 2 or i == 3:
    fit = a(random_state = 2022).fit(X_train,y_train)
  else:
    fit = a().fit(X_train,y_train)
  pred = fit.predict(X_test)
  r2 = r2_score(y_test, pred)
  mse = mean_squared_error(y_test, pred)

  P.append(pred)
  print(f'[ {a.__name__} ]\n - R2: {r2:.4f} | MSE : {mse:.4f} \n'  )

#  DataFrame 구조로 만들기
df = pd.DataFrame({
    'y_test':y_test, 'LR':P[0], 'SVR':P[1], 'RFR':P[2], 'XGB':P[3]
})
df.head()
```
