# 다중회귀 - 보스톤 주택 가격 

```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# 경고창 무시 
import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_boston # 보스턴 데이터 
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['PRICE'] = boston.target
df.head()
```

- Feature 별 산점도와 선형 회귀선 

```python
sns.regplot(x='CRIM', y='PRICE', data=df);
```

```python
# 반복문을 통한 다중 그래프 그리기
_, axs = plt.subplots(3, 4, figsize=(16,10)) # 3행 4열 총 12개 그래프 
for i, feature in enumerate(df.columns[1:-1]):
  row, col = i //4 , i % 4 # index 값이 4의 배수인 경우 = 1
  sns.regplot(x=feature, y='PRICE', data=df, ax=axs[row][col])           
```

- Feature 별 결정계수(R squared)

```python
from sklearn.linear_model import LinearRegression
```

```python
for feature in df.columns[:-1]:
  lr = LinearRegression()
  X = df[feature].values.reshape(-1,1)  # 행 여러개, 열 1개 
  lr.fit(X, boston.target)
  score =lr.score(X, boston.target)
  print(f'{feature}:\t{score:.4F}')
```

### 회귀는 꼭 선형회귀를 써야 하는가?
- `LinearRegression`, `Decision Tree`, `SVM`,`RandomForest`, `XGboost`

```python
from sklearn.metrics import r2_score , mean_squared_error
```

#### 1. `LinearRegression`

```python
pred_lr = lr.predict(X_test) 
r2_lr = r2_score(y_test, pred_lr)
mse_lr = mean_squared_error(y_test, pred_lr)
```

#### 2. `Decision Tree`

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=2022)
dtr.fit(X_train, y_train)
```

```python
pred_dt = dtr.predict(X_test)                # 예측
r2_dt = r2_score(y_test, pred_dt)            # r2_score
mse_dt = mean_squared_error(y_test, pred_dt) # mean_squared_error
```

#### 3. `SVM`

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
```

```python
pred_sv = dtr.predict(X_test)                # 예측
r2_sv = r2_score(y_test, pred_sv)            # r2_score
mse_sv = mean_squared_error(y_test, pred_sv) # mean_squared_error
```

#### 4. `RandomForest`

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state = 2022)
rfr.fit(X_train, y_train)                    # 학습
pred_rf = rfr.predict(X_test)                # 예측
r2_rf = r2_score(y_test, pred_rf)            # r2_score
mse_rf = mean_squared_error(y_test, pred_rf) # mean_squared_error
```

#### 5. `XGBoost`

```python
from xgboost import XGBRegressor
xgr = XGBRegressor(random_state = 2022)
xgr.fit(X_train, y_train)                    # 학습 
pred_xg = xgr.predict(X_test)                # 예측
r2_xg = r2_score(y_test, pred_xg)            # r2_score
mse_xg = mean_squared_error(y_test, pred_xg) # mean_squared_error
```

#### 6. 비교

```python
print('LR \t Dt \t SVM \t RF \t XG')
print(f'{r2_lr :.4f}\t{r2_dt :.4f}\t{r2_sv :.4f}\t{r2_rf :.4f}\t{r2_xg :.4f}')
print(f'{mse_lr :.4f}\t{mse_dt :.4f}\t{mse_sv :.4f}\t{mse_rf :.4f}\t{mse_xg :.4f}')
```
