- `BMI` vs `target`
 - train/test dataset split (test_size=0.1)

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
    df.bmi.values.reshape(-1,1), diabetes.target, test_size=0.1, random_state=2022
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape # 거의 train:test =  9:1에 가까운 값
```

```python
lr  = LinearRegression()
lr.fit(X_train, y_train)  # 학습
```

```python
# 계수, 절편, 잔차제곱의 합 
lr.coef_, lr.intercept_, lr._residues
```

```python
# R squared
from sklearn.metrics import r2_score
pred = lr.predict(X_test)
r2_score(y_test, pred)
```

```python
lr.score(X_test, y_test)
```

```python
# Mean Squared Error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
mse, rmse
```
