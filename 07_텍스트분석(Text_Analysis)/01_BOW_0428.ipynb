# BOW (Bag Of Words)


```python
 from sklearn.feature_extraction.text import CountVectorizer

text = "The SMS Spam Collection is a set of SMS tagged messages that have been collected for SMS Spam research."

cvect = CountVectorizer()
cvect.get_params()
```

```python
output = cvect.fit_transform([text]) # 꼭 리스트로 변환해서 사용 !
output
```

```python
output.toarray()
```

```python
cvect.vocabulary_
```

- 불용어(stop word) 처리
 - 자체적으로 불용어 지정한 제거 방법
 - Scikit-learn에서 제공하는 불용어 제거 방법
 - NLTK 불용어 사용 

```python
# 방법 1) 자체 제거 
cvect = CountVectorizer(stop_words=['is', 'a', 'the']) # 관사 제거
print(cvect.fit_transform([text]).toarray())
print(cvect.vocabulary_)
```

```python
# 방법 2) Scikit-learn에서 제공하는 불용어 제거
cvect = CountVectorizer(stop_words='english') 
print(cvect.fit_transform([text]).toarray())
print(cvect.vocabulary_)
```

```python
# 방법 3) NLTK 불용어 사용 
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
sw = stopwords.words('english')
len(sw), type(sw)
```

```python
cvect = CountVectorizer(stop_words = sw)  # stop_words = sw 
print(cvect.fit_transform([text]).toarray())
print(cvect.vocabulary_) # 딕셔너리 형태
```

### 2. N-gram


```python
text2 = ['I work at google.', 'I google at work .']

cvect = CountVectorizer()
print(cvect.fit_transform(text2).toarray()) # text2 리스트 형태여서 그냥 사용  
print(cvect.vocabulary_) # 딕셔너리 형태
```

```python
# ngram_range=(1,2)
cvect = CountVectorizer(ngram_range=(1,2))
print(cvect.fit_transform(text2).toarray())  
print(cvect.vocabulary_) # 딕셔너리 형태
# ngram 문제점 -> 수가 늘어남. 처리 시간 증가!! 
```

### 3. TfidfVectorizer(Term Frequency, Inverse Document Frequency)
- TF-IDF는 모든 문서에서 자주 등장하는 단어는 중요도가 낮다고 판단하며, 특정 문서에서만 자주 등장하는 단어는 중요도가 높다고 판단

```python
text3 = ["The SMS Spam Collection is a set of SMS tagged messages that have been collected for SMS Spam research.",
         "It contains one set of SMS messages in English of 5,574 messages, tagged acording being ham (legitimate) or spam."]
```

```python
# TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tvect = TfidfVectorizer(stop_words = 'english')
print(tvect.fit_transform(text3).toarray())
print(tvect.vocabulary_)
```

```python
# CountVectorizer (위와 비교 위헤)
from sklearn.feature_extraction.text import CountVectorizer
cvect = CountVectorizer(stop_words='english')
print(cvect.fit_transform(text3).toarray())
print(cvect.vocabulary_)
```
